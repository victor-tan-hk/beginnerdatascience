```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Visualise categorical data

- bar chart (preferred)
- pie chart

```python
fname = 'data/tips.csv'
tips = pd.read_csv(fname)
tips
```

```python
set(tips.day)
```

```python
days = ['Thur', 'Fri', 'Sat', 'Sun']
freq = [list(tips.day).count(d) for d in days]
freq
```

```python
list(tips.day).count('Sun')
```

```python
sum(tips.day == 'Sun')
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.bar(x=days, height=freq)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))
# days = ['Thur', 'Fri', 'Sat', 'Sun']
highlight = [0, 0.2, 0, 0]
ax.pie(x=freq, labels=days, explode=highlight)
plt.show()
```

# Numerical data

- distribution -> histogram
- outliers -> boxplot
- relationship -> scatter plot

## Histogram

```python
fig, ax = plt.subplots(1,2, figsize=(3,2))
ax[0].hist(tips.total_bill)
ax[1].hist(tips.tip)
```

```python
fig, ax = plt.subplots(1,2, figsize=(3,2))
ax[0].hist(tips.total_bill, bins=30)
ax[1].hist(tips.tip, bins=30)
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(3,2))
ax[0].hist(tips.total_bill, bins=[0,10,20,30,40,50])
ax[1].hist(tips.tip, bins=np.linspace(0,10, 11))
plt.show()
```

```python
np.linspace(0, 1, 11)
```

```python
print(set(tips.sex))
mask = tips.sex=='Female'
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.hist(tips.total_bill[mask], bins=20, color='blue', alpha=0.3)
ax.hist(tips.total_bill[~mask], bins=20, color='red', alpha=0.3)
plt.show()
```

```python
mask = tips.sex=='Female'
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.hist(tips.tip[mask], bins=20, color='blue', alpha=0.3)
ax.hist(tips.tip[~mask], bins=20, color='red', alpha=0.3)
plt.show()
```

## Boxplot

```python
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.boxplot(tips.total_bill)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.boxplot(x=tips[['total_bill', 'tip']])
plt.show()
```

## Scatter plot

```python
fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.scatter(tips.total_bill, tips.tip, s=2)
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
mask = tips.sex == 'Female'

fig, ax = plt.subplots(1,1, figsize=(3,2))
ax.scatter(tips.total_bill[mask], tips.tip[mask], s=5, color='blue', alpha=0.5)
ax.scatter(tips.total_bill[~mask], tips.tip[~mask], s=5, color='red', alpha=0.5)
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['sl','sw','pl','pw','target']
iris
```

```python
mask_setosa = iris['target']=='setosa'
mask_versicolor = iris['target']=='versicolor'
mask_virginica = iris['target']=='virginica'
```

```python
fig, ax = plt.subplots(1,1, figsize=(4,3))
ax.scatter(iris.sl[mask_setosa], iris.pl[mask_setosa],
           s=5, c='red', alpha=0.4)
ax.scatter(iris.sl[mask_versicolor], iris.pl[mask_versicolor],
           s=5, c='green', alpha=0.4)
ax.scatter(iris.sl[mask_virginica], iris.pl[mask_virginica],
           s=5, c='blue', alpha=0.4)
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(iris)
plt.show()
```
